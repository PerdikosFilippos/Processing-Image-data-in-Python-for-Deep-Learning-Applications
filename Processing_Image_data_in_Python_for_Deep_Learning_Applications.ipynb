#Resizing th image using the Pillow library

```python
from PIL import Image
```

```python
img = Image.open("/content/Dog.jpg")
img_resized = img.resize((200,200))
```

```python
#save the new resized image
img_resized.save("dog_img_resized.jpg")
```

```python
#Display image from numpy array
img_res = mpimg.imread("/content/dog_img_resized.jpg")
img_res_plot = plt.imshow(img_res)
plt.show()
```

```python
img_res.shape
```

#Convert RGB images to grayscale

```python
#importing OpenCV library
import cv2
```

```python
img = cv2.imread("/content/dog_img_resized.jpg")
```

```python
type(img)
```

```python
img.shape
```

```python
#Create grayscale
grayscale_image = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
```

```python
grayscale_image.shape
```
